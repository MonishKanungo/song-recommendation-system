```python
from sklearn.cluster import KMeans
import numpy as np

def clustering (column_name):
    # column_name = "Danceability"
    data1 = data[column_name].values.reshape(-1, 1)

    # Choose the number of clusters
    num_clusters = 3

    # Create a KMeans instance and fit it to the data
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(data1)

    # Get cluster labels assigned to each data point
    cluster_labels = kmeans.labels_

    # Print the cluster labels
    data[column_name]=cluster_labels
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
cv.fit_transform(song["tag"])
vector=cv.fit_transform(song["tag"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarityMatrix=cosine_similarity(vector)
print(similarityMatrix)
```
